# **1 - Интерактивность**

```
from traitlets.utils import descriptions
from ipywidgets import interactive
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np

def my_days_parametrs(eat = 1, sleep = 5, gaming = 3, studying = 2, programming = 4, working = 5, reading = 0.5, chating = 1.5, watching = 1):
  doies_names = ["Ем", "Сплю", "Играю", "Учусь", "Программирую", "Работаю", "Читаю статьи", "Сижу в соц.сетях", "Смотрю видео"]
  doies_times = [eat, sleep, gaming, studying, programming, working, reading, chating, watching]
  free_time = 24 - sum(doies_times)
  doies_names.append ("Свободное время")
  doies_times.append(free_time)

  explode = [0]*len(doies_times)
  explode[-1] = 0.1

  plt.figure(figsize = (16,8))
  cmap = plt.get_cmap("tab20b")
 
  plt.subplot(1,2,1)
  plt.pie(doies_times,
          explode=explode,
          labels=doies_names,
          autopct='%1.1f%%',
          shadow=True,
          startangle=90)
  
  if free_time < 0:
    plt.title("Вы ввели более 24 часов")
  else:
    plt.title("При таком расписании у меня %d свободных часов "% (free_time))

  plt.subplot(1,2,2)
  patches, texts = plt.pie([1+5,3+2,4+5,0.5+1.5,1+free_time],
                           colors = cmap(np.array([0, 5])),
                           radius = 1,
                           startangle=90,
                           wedgeprops=dict(width=0.3, edgecolor='w'))

  plt.pie(doies_times,
          labels=doies_names,
          radius = 0.7,
          colors = cmap(np.array([1,2,6,7,1,2,6,7,1,2])), 
          autopct='%1.1f%%',
          startangle=90,
          wedgeprops=dict(width=0.3, edgecolor='w'))
  
  plt.legend(patches, ['Дела', 'Отдых'], loc="best")
  plt.show()

w = interactive(my_days_parametrs, eat = (0,24), sleep = (0,24), gaming = (0,24), studying = (0,24), programming = (0,24), working = (0,24), reading = (0,24), chating = (0,24), watching = (0,24))
display(w)
```
