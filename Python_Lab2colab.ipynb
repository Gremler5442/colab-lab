```
from traitlets.utils import descriptions
from ipywidgets import interactive
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np

def my_days_parametrs(eat = 1, sleep = 5, gaming = 3, studying = 2, programming = 4, working = 5, reading = 0.5, chating = 1.5, watching = 1):
  doies_names = ["", "", "", "", "", "", " ", "  .", " "]
  doies_times = [eat, sleep, gaming, studying, programming, working, reading, chating, watching]
  free_time = 24 - sum(doies_times)
  doies_names.append (" ")
  doies_times.append(free_time)

  explode = [0]*len(doies_times)
  explode[-1] = 0.1

  plt.figure(figsize = (16,8))
  cmap = plt.get_cmap("tab20b")
 
  plt.subplot(1,2,1)
  plt.pie(doies_times,
          explode=explode,
          labels=doies_names,
          autopct='%1.1f%%',
          shadow=True,
          startangle=90)
  
  if free_time < 0:
    plt.title("   24 ")
  else:
    plt.title("     %d   "% (free_time))

  plt.subplot(1,2,2)
  patches, texts = plt.pie([1+5,3+2,4+5,0.5+1.5,1+free_time],
                           colors = cmap(np.array([0, 5])),
                           radius = 1,
                           startangle=90,
                           wedgeprops=dict(width=0.3, edgecolor='w'))

  plt.pie(doies_times,
          labels=doies_names,
          radius = 0.7,
          colors = cmap(np.array([1,2,6,7,1,2,6,7,1,2])), 
          autopct='%1.1f%%',
          startangle=90,
          wedgeprops=dict(width=0.3, edgecolor='w'))
  
  plt.legend(patches, ['', ''], loc="best")
  plt.show()

w = interactive(my_days_parametrs, eat = (0,24), sleep = (0,24), gaming = (0,24), studying = (0,24), programming = (0,24), working = (0,24), reading = (0,24), chating = (0,24), watching = (0,24))
display(w)
```

```
import matplotlib.pyplot as plt
letters = [" -    ,      .",
" -         .",
" -   ,    .",
" -         .",
" -   ,      .",
" -    ,    .",
" -       .",
" -    ,    .",
" -    ,  ,    .",
" -      ,      .",
" -  , ,   .",
" -   ,      .",
"    ,         .",
" -   ,    .",
" -     ,    .",
" -     ,    .",
" -   ,  ,    .",
" -      ,       .",
" - ,   ,      .",
" -    ,   .",
" -   ,    .",
"    ,    ",
"     ,     .",
" -         ."]

name = "\u0410\u043B\u0435\u043A\u0441\u0430\u043D\u0434\u0440" #@param {type:"string"}
result = []

for ch in name:
  if ch == " ":
    continue
  for it in letters:
    if it[0] == ch.upper():
      result.append(it)

sorted = []
for i in result:
  if i not in sorted:
    sorted.append(i)

sorted
```

```
graph = []
for ch in name:
  if ch == ' ':
    continue
  graph.append(ch.upper())

result = {i: graph.count(i) for i in graph}

plt.figure(figsize=(16,16))
plt.bar(result.keys(), result.values())
plt.title(' ')
plt.xlabel('')
plt.ylabel('')
plt.gcf().autofmt_xdate()
```

```
import matplotlib.pyplot as plt
import numpy as np
energy = 3 #@param {type:"slider", min:0, max:10, step:1}
positive = 3 #@param {type:"slider", min:0, max:10, step:1}
pleasure = 2 #@param {type:"slider", min:0, max:10, step:1}
folks = 4 #@param {type:"slider", min:0, max:10, step:1}

p = energy + positive
e = pleasure
h = folks

happines = p + (5*e) + (3*h)

array = [energy, positive, pleasure, folks]
array_labels = ["","","",""]

plt.figure(figsize = (9, 9))
plt.pie(array,
        labels=array_labels, 
        autopct='%1.1f%%',
        startangle=90)
plt.title("  : %d"% (happines))

plt.show()
```
