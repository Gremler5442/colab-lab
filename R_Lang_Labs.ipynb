# **ПОСТРОЕНИЕ HEATMAP**

```r
#ПОЛУЧАЕМ ДАННЫЕ, ОБРАБАТЫВАЕМ CSV
data <- read.csv("http://datasets.flowingdata.com/ppg2008.csv", sep=",", header = TRUE)
```

```r
#ВЫВОДИМ ПЕРВЫЕ 5 СТРОК ТАБЛИЦЫ
data[1:5,]
```

```r
#СОРТИРУЕМ ПО FPG ПО УБЫВАНИЮ
sorted_fgp <- data[order(data$FGP, decreasing = TRUE),]
sorted_fgp[1:5,]
```

```r
#СОРТИРУЕМ ПО PTS ПО ВОЗРАСТАНИЮ
sorted_pts <- data[order(data$PTS, decreasing = FALSE),]
sorted_pts[1:5,]
```

```r
#ПЕРЕНОС КОЛОНКИ ИМЕН В НАЗВАНИЯ СТРОК
row.names(data) <- data$Name
data <- data[,2:20]
data
```

```r
#СТРОИМ МАТРИЦУ ДЛЯ HEATMAP
data_matrix <- data.matrix(data)
```

```r
#СТРОИМ HEATMAP
data_heatmap <- heatmap(data_matrix, Rowv = NA, Colv = NA, col = cm.colors(256), scale = "column", margins = c(5, 10))
```

```r
#СКАЧИВАЕМ И УСТАНАВЛИВАЕМ RColorBrewer
if (!require("RColorBrewer")) install.packages("RColorBrewer")
library(RColorBrewer)
```

```r
#ПОЛУЧАЕМ ПАЛИТРУ ЦВЕТОВ
brewer.pal.info
```

```r
#КАСТОМИЗИРУЕМ HEATMAP
data_heatmap <- heatmap(data_matrix, Rowv = NA, Colv = NA, col = brewer.pal(9, "Purples"), scale = "column", margins = c(5, 10))
```

# **ПОСТРОЕНИЕ РАДАРА CRIMES**

```r
crime_radar <- read.csv("http://datasets.flowingdata.com/crimeRatesByState2005.tsv", sep="\t", header = TRUE)
```

```r
row.names(crime_radar) <- crime_radar$state
crime_radar <- crime_radar[,2:7]
stars(crime_radar, flip.labels = FALSE, key.loc = c(15, 1.5))
```

```r
stars(crime_radar, flip.labels = FALSE, key.loc = c(15, 1.5), full = FALSE)
```

```r
stars(crime_radar, flip.labels = FALSE, key.loc = c(15, 1.5), draw.segments = TRUE)
```
