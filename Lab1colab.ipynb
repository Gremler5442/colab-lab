# Надписи столбцов


````
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
````

````
legend_elements = [Patch(facecolor='#a5a58d', label='Category 2'),
                   Patch(facecolor='#cb997e', edgecolor='lightgrey', label='Category 1')]
plt.bar(['first', 'second', 'third'], [2, 4, 6], color = ['#a5a58d', '#a5a58d', '#cb997e'])
plt.legend(handles=legend_elements)
ax = plt.gca()
# plt.ylim(top=8)

total = sum([2, 4, 6])
ax.text(-0.46, 4.9, f'Всего: {total} (100%)')
for i, v in enumerate([2, 4, 6]):
    h = v*1.05
    ax.text(i-0.2, h, f'{v} ({v / total * 100.0:.1f} %)')

ax.get_yaxis().set_ticks([])
ax.set(frame_on=False)
plt.tight_layout()
plt.title('Title', fontsize = 16)
plt.show()
````
