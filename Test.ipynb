# **1 - Управляем своей занятостью динамически**

Как много дел! А хватит ли у тебя времени, чтобы работать в нашем отделе? Нам важно чтоб у вас было хотя бы 6 часов свободного времени.
Передлааем нашу программу так, чтобы была возможность в интерактивном режиме подкрутить время и понять, где мы можем сэкономить времени, чтобы вы успевали у нас работать и при этом не забывали отдыхать.

Поэксперементируйте с параметрами.


````python
from ipywidgets import interactive
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt

# Пишем функцию, которая строит график наших дел.
# Аргументами функции являются наши дела, их значения - время которое мы на них затрачиваем
def my_days_parametrs(sleep = 10, eat = 3, learn = 6, gaming = 1):
  doies_names = ['сплю', 'ем', 'учусь', 'играю'] #названия наших дел на русском
  doies_times = [sleep, eat, learn, gaming] # время которое дела занимают (просто добавляем в массив значения)
  free_time = 24 - sum(doies_times) #рассчитываем количество свободного времени
  doies_names.append ("Свободное время")
  doies_times.append(free_time)

  explode = [0]*len(doies_times) #этот параметр обозначает на сколько один кусок диаграммы выдвинут
  explode[-1] = 0.1 #настраиваем, чтобы выдвигалось Свободное время

  
  plt.figure(figsize = (10,10))
  plt.pie(doies_times, explode=explode, labels=doies_names, autopct='%1.1f%%',
          shadow=True, startangle=90)
  
  plt.title("При таком расписании у меня %d свободных часов "% (free_time))

  plt.show()
btn = widgets.Button(
    value=False,
    description='Click me',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Description',
    icon='check' # (FontAwesome names without the `fa-` prefix)
)
w = interactive(my_days_parametrs,  sleep = (0,24), eat = (0,24), learn = (0,24), gaming = (0,24))
display(w, btn)
````

# **2 - Задача на манипуляторы**
Для того, чтобы территория процветала и среда обитания населения города была комфортной, а люди счастливыми, необходимо внедрить несколько умных решений. Каждое умное решение имеет оценку, указывающую на степень одобрения его жителями района - баллы счастья. Администрация может :
* разместить пункты сбора раздельного мусора (waste) - 1000  баллов за один пункт, 
* открыть общепиты/фудкорты (cafe) - 3000 баллов за одно предприятие, 
* настроить вай-фай на площадках (wifi) - 1500 баллов за одну площадку 
* установить кнопки экстренной помощи (helps) -600 баллов за один двор.

Кроме того, умные решения требуют финансовых затрат, которые у нас ограничены: городом администрации района выделено - 170 000 условных денежных единиц (уде), при этом:
* на пункт мусора затрачивается 7000 уде., 
* на предприятие общественного питания - 5000 уде, 
* на wi-fi точку - 30000 уде, 
* на копку помощи - 10000 уде .

Определите оптимальное **количество** умных решений достаточных для обеспечения комофртной среды обитания района и укладывающихся в бюджет. Известно, что администрация намерена реализовать все умные решения, а это значит, что количество размещаемых объектов (кафе, точки wi-fi, пункты сбора раздельного мусора, кнопки экстренной помощи) не могут быть равны нулю.

Параметр необходимого уровня комфорта устанавливается таким образом, чтобы его значение было не ниже макимального показателя в других районах города (жители которых довольны обсутройством и жизнееобеспечением территории): 39020, 40000, 38800, 39800.

````python
from ipywidgets import Dropdown, Checkbox, HBox, VBox, Text, Button
from IPython.display import display, clear_output

def counts_of_points(waste = 10, cafe = 10, wifi = 10, helps = 10):

    #стоимость
    mwaste = waste*7000
    mcafe = cafe*5000
    mwifi = wifi*30000
    mhelp = helps*10000
    spends = [mwaste, mcafe, mwifi, mhelp]
    maxspend = 170000
    title_1 = "Текущие траты %d!"%sum(spends) if sum(spends) < maxspend else "Текущие траты %d! Превышают лимит 170-000"%sum(spends) 

    #баллы счастья
    max_happy = waste*1000 + wifi*1500 + cafe*3000 + helps*600

    other_centers = [39020, 40000, 38800, 39800, max_happy ]

    title_2 = "Счастья слишком мало!" if max(other_centers[:-1]) > other_centers[-1] else "Счастья сколько надо!"

    plt.figure(figsize = (10, 4))
    plt.subplot(1,2,1)
    plt.pie(spends, labels=spends, autopct='%1.1f%%',
        shadow=True, startangle=90)
    plt.title(title_1)

    plt.subplot(1,2,2)

    plt.pie(other_centers, labels=other_centers, autopct='%1.1f%%',
        shadow=True, startangle=90)
    plt.title(title_2)
    plt.tight_layout()
    plt.show()

import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
from IPython.display import display

w = interactive(counts_of_points, waste = 5, cafe = 5, wifi = 5, helps = 5)
display(w)
````

**Работа с функцией .find_all()**

В переменной soup содержится структура html. Чтобы получить нужные данные, нужно обращаться по тегу и аттрибуту. Для этого воспользуемся функцией `find_all`

Сначала в функции указывается тег в кавычках, далее, если необходимо, через запятую в фигурных скобках - аттрибуты в виде: `{'название аттрибута':'значение аттрибута'}`

Так как функция `find_all` возвращает список тегов и то, что внутри них, нам нужно пройтись по каждому элементу этого списка и получить то, что нужно:

|Часть   | Код    |
|:-------|:-------|
| тег    |`.название_тега`|
| аттрибут | `['название аттрибута']`|
| соддержимое | `.text`|

# **УДАЛИТЬ: 3 - Интерактивная фильтрация датасетов**

Ниже написан код, реализующий интерактивную фильтрацию данных. Это значит, что для числовых значений реализованы манипуляторы. Для выбора поля сортировки - выпадающий список. Для определения направления сортировки - однозначный выбор.

Этот код вы можете использовать для других параметров, других датафреймов, добавлять нужные поля и т.д.

Дополнительный материал, применение интерактивности в практике, на примере прогноза демографии: https://colab.research.google.com/drive/1e2zV42Bc4YZsCVy2TlVxDak4verI-uG-?usp=sharing 

````python
import pandas as pd
import io
import requests
import numpy as np
import matplotlib.pyplot as plt
def read_csv(url):
  s=requests.get(url).content
  return(pd.read_csv(io.StringIO(s.decode('utf-8'))))
````

Подробнее о работе с таблицами (датафреймами): https://colab.research.google.com/drive/1E6Iqdy0iumta6M1tkLHWKvEK3L9B8Tsq?usp=sharing

````python
dogs = read_csv('https://raw.githubusercontent.com/yupest/nti/master/НТИ-2021/dogs.csv')
````

````python
import ipywidgets as widgets
from IPython.display import display, clear_output
from ipywidgets import interact, interactive, fixed, interact_manual
````

````python
#функция отображения параметров и таблицы
def show_df(weight, height, life, sorts, asc, count):
  #очистить экран
  # clear_output()
  #отобразить параметры и таблицу по фильтрам, срезу и сортировке
  display(dogs.sort_values(sorts, ascending = asc).loc[
        (dogs["Вес мин"]>=weight[0]) & 
        (dogs["Вес макс"]< weight[1]) &
        (dogs["Рост мин"]>= height[0]) &
        (dogs["Рост макс"]< height[1]) &
        (dogs["Продолжительность жизни мин"]>= life[0]) &
        (dogs["Продолжительность жизни макс"]< life[1]) 
        ][count[0]:count[1]])
# создание манипуляторов
weight = widgets.IntRangeSlider(
    min=dogs['Вес мин'].min(),
    max=dogs['Вес макс'].max(),
    step=1,
    continuous_update=False,
    description='Вес:'
)
height = widgets.IntRangeSlider(
    min=dogs['Рост мин'].min(),
    max=dogs['Рост макс'].max(),
    step=1,
    continuous_update=False,
    description='Рост:'
)
life = widgets.IntRangeSlider(
    min=dogs['Продолжительность жизни мин'].min(),
    max=dogs['Продолжительность жизни макс'].max(),
    step=1,
    continuous_update=False,
    description='Продолжительность жизни:'
)
#Выпадающий список со значениями - названиями полей
sorts = widgets.Dropdown(
    options = list(dogs.columns),
    description = 'Сортировать по полю:'
)
#направление сортировки по возрастанию или нет
asc = widgets.RadioButtons(
    options = [("Да", True), ('Нет', False)],
    description = 'По возрастанию'
)
#определение среза от и до
count = widgets.IntRangeSlider(
    min=0,
    max=dogs.shape[0],
    step=1,
    continuous_update=False,
    description='Срез:'
)
#для изменения каждого параметра должны обновляться выходные
w = interactive(show_df, weight = weight, height = height, life = life, sorts = sorts, asc = asc, count = count)
#сразу отображаем все параметры и таблицу
display(w)
````
